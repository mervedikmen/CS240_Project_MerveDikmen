```python
from __future__ import print_function

import numpy as np
import pandas as pd

import thinkplot
import thinkstats2

%matplotlib inline
```

# Section 2 - Data Reading

```python
data = pd.read_csv('basketball_player_allstar.csv',encoding="ISO-8859-1", delimiter=";") #reading database

minutes = data['minutes']    #minutes played
points = data['points']      #points gathered

print (minutes)
print (points)
```

# Section 3 - Descriptive Statistics, Histograms, PMF & CDF

```python
#Descriptive Statisctics for Munites

max_m = minutes.max()     #maximum minutes
min_m = minutes.min()     #minimum minutes
std_m = minutes.std()     #standard deviation of minutes
mean_m = minutes.mean()   #mean of minutes
median_m = minutes.median()     #median minutes
mode_m = minutes.mode()     #mode of minutes

print ('Descriptive Statisctics for Munites')
print ('Maximum Minute-> ' + str(max_m))
print ('Minimum Minute-> ' + str(min_m))
print ('Standard Deviation Of Minutes-> ' + str(std_m))
print ('Mean of Minutes-> ' + str(mean_m))
print ('Median of Minutes-> ' + str(median_m))
print ('Mode of Minutes-> ' + str(mode_m))
print ()

#Descriptive Statisctics for Points

max_p = points.max()     #maximum point
min_p = points.min()     #minimum point
std_p = points.std()     #standard deviation of points
mean_p = points.mean()     #mean of points
median_p = points.median()     #median of points
mode_p = points.mode()     #mode of points

print ('Descriptive Statisctics for Points')
print ('Maximum Point-> ' + str(max_p))
print ('Minimum Point-> ' + str(min_p))
print ('Standard Deviation of Points-> ' + str(std_p))
print ('Mean of Points-> ' + str(mean_p))
print ('Median of Points-> ' + str(median_p))
print ('Mode of Points-> ' + str(mode_p))
```

```python
#Histogram
#Histogram shows the distribution of data of minutes and points. 
#Point values is higher than minutes values generally.

minutes_histogram =thinkstats2.Hist(minutes ,label='Minutes')
points_histogram = thinkstats2.Hist(points,label='Points')
thinkplot.Hist(minutes_histogram, color='Green',  width=0.50, align='right')
thinkplot.Hist(points_histogram, color='Orange',  width=0.50, align='left') 
thinkplot.Show(xlabel='Minutes and Points Values', loc='upper right', title='Histogram For Minutes and Points')
```

```python
#PMF
#While histogram shows the minutes and points values, PMF gives the probability of each value.

minutes_pmf = thinkstats2.Pmf(minutes,label='Minutes')
points_pmf = thinkstats2.Pmf(points,label='Points')
thinkplot.PrePlot(2)
thinkplot.Pmf(minutes_pmf, color='Purple', align='right')
thinkplot.Pmf(points_pmf, color='Pink', align='right') 
thinkplot.Show(xlabel='Minutes and Points Values', ylabel='PMF', loc='upper right', title='PMF of Minutes and Points')
```

```python
#CDF
#Cdf gives us the each value's percentile rank and it is calculated as adding the probabiity of each value to the next one. 

minutes_cdf= thinkstats2.Cdf(minutes)
points_cdf = thinkstats2.Cdf(points)
thinkplot.Cdf(minutes_cdf, color='Blue',label='Minutes')
thinkplot.Cdf(points_cdf, color='Yellow',label='Points') 
thinkplot.Show(xlabel='Minutes and Points Values', ylabel='CDF', loc='upper right', title='Cdf of Minutes and Points')
```

# Section 4 - Modelling The Data

```python
#Normal Distribution

std_p = points.std()
mean_p = points.mean()

pdf = thinkstats2.NormalPdf(mean_p, std_p)

title1='Normal distribution Model For Points'
thinkplot.Config(xlabel='Points', title=title1, legend  = True)
thinkplot.Pdf(pdf, color='green')
print ('The normal distribution model for the points data is shown above with mean ' + str(mean_p) + ' and std ' + str(std_p))
```

# Section 5 - Correlation Between Minutes And Points

```python
#Vizualizing the correlation between minutes and points
#Scatter plot is the best way to show the relationship between two variables.
#In this scatter plot, it is shown that as minutes increases, the points also increases. 

thinkplot.Scatter(minutes, points, alpha=0.2,color='green') 
thinkplot.Config(xlabel='Played Minutes', ylabel='Gained Points', title= 'Scatter Plot For Minutes versus Points', legend=True)
```

# Section 6 - Hypotesis Testing for minutes and points

```python
#To test the null hypothesis, hypothesis testing is the important way.

class BasketballTesting(thinkstats2.HypothesisTest):
    
    def TestStatistic(self, data):        #with the test statistic, the effect's size can be measured.
        value1, value2 = data
        mean_a = value1.mean()
        mean_b = value2.mean()
        test_statistic = abs(mean_a - mean_b)     #the test statistic is the absolute value of differences of values' mean.
        return test_statistic
    
    def MakeModel(self):      #combine the values as self.pool
        value1, value2 = self.data
        self.a, self.b = len(value1), len(value2)
        self.pool = np.hstack((value1, value2))
        
    def RunModel(self):      #simulates the null hypotesis by shuffling 
        np.random.shuffle(self.pool)
        data = self.pool[:self.a], self.pool[self.b:]
        return data

data = points, minutes
hypothesis_testing = BasketballTesting(data)
pvalue = hypothesis_testing.PValue()      #finding p-value
print ('P-value for minutes and points data: ' + str(pvalue))
```
